```r
library(ggplot2)
library(dplyr)
library(stringr)
library(lubridate)
library(gridExtra)
library(readr)
library(urbnmapr)
library(viridis)
library(ggthemes)
library(ggsci)
library(tidyr)
```

```r
fp       <- '/pool001/mfzhao/'
weather  <- read_csv(str_c(fp, 'PROCESSED_DATA/region_weather_2020.csv'))
counties <- get_urbn_map(map = "counties", sf = TRUE)
```

```r
counties %>%
    left_join(weather %>%
              rename(county_fips = key) %>%
              filter(DATE == as.Date('2020-5-21'))) %>%
    ggplot(aes(fill = PRCP)) +
    geom_sf(size = 0) +
    scale_fill_material('blue') +  
    xlab('') + 
    ylab('') +
    ggtitle('May 21, 2020') + 
    labs(color = "Outcome") +
    theme_light() +
    theme(text = element_text(size=15),
          plot.title = element_text(vjust = -5, hjust = .75),
          strip.text.y = element_blank(),
          panel.border=element_blank(),
          panel.grid.major=element_blank(), ## element_line(colour = "white")
          panel.grid.minor=element_blank(), ## element_line(colour = "white", size = 0.25)
          axis.ticks=element_blank(),       ## element_line(colour = "grey20")
          axis.text=element_blank(),        ## element_text(size = rel(0.8), colour = "grey30")
          axis.title=element_blank(),      ## axis.title.x = element_text(
          plot.margin= grid::unit(c(0, 0, 0, 0), "in"),
          legend.position = 'None') -> prcp1

counties %>%
    left_join(weather %>%
              rename(county_fips = key) %>%
              filter(DATE == as.Date('2020-5-22'))) %>%
    ggplot(aes(fill = PRCP)) +
    geom_sf(size = 0) +
    scale_fill_material('blue') +  
    xlab('') + 
    ylab('') +
    ggtitle('May 22, 2020') + 
    labs(color = "Outcome") +
    theme_light() +
    theme(text = element_text(size=15),
          plot.title = element_text(vjust = -5, hjust = .75),
          strip.text.y = element_blank(),
          panel.border=element_blank(),
          panel.grid.major=element_blank(), ## element_line(colour = "white")
          panel.grid.minor=element_blank(), ## element_line(colour = "white", size = 0.25)
          axis.ticks=element_blank(),       ## element_line(colour = "grey20")
          axis.text=element_blank(),        ## element_text(size = rel(0.8), colour = "grey30")
          axis.title=element_blank(),      ## axis.title.x = element_text(
          plot.margin= grid::unit(c(0, 0, 0, 0), "in"),
          legend.position = 'None') -> prcp2

counties %>%
    left_join(weather %>%
              rename(county_fips = key) %>%
              filter(DATE == as.Date('2020-5-23'))) %>%
    ggplot(aes(fill = PRCP)) +
    geom_sf(size = 0) +
    scale_fill_material('blue') +  
    xlab('') + 
    ylab('') +
    ggtitle('May 23, 2020') + 
    labs(color = "Outcome") +
    theme_light() +
    theme(text = element_text(size=15),
          plot.title = element_text(vjust = -5, hjust = .75),
          strip.text.y = element_blank(),
          panel.border=element_blank(),
          panel.grid.major=element_blank(), ## element_line(colour = "white")
          panel.grid.minor=element_blank(), ## element_line(colour = "white", size = 0.25)
          axis.ticks=element_blank(),       ## element_line(colour = "grey20")
          axis.text=element_blank(),        ## element_text(size = rel(0.8), colour = "grey30")
          axis.title=element_blank(),      ## axis.title.x = element_text(
          plot.margin= grid::unit(c(0, 0, 0, 0), "in"),
          legend.position = 'None') -> prcp3

counties %>%
    left_join(weather %>%
              rename(county_fips = key) %>%
              filter(DATE == as.Date('2020-5-24'))) %>%
    ggplot(aes(fill = PRCP)) +
    geom_sf(size = 0) +
    scale_fill_material('blue') +  
    xlab('') + 
    ylab('') +
    ggtitle('May 24, 2020') + 
    labs(color = "Outcome") +
    theme_light() +
    theme(text = element_text(size=15),
          plot.title = element_text(vjust = -5, hjust = .75),
          strip.text.y = element_blank(),
          panel.border=element_blank(),
          panel.grid.major=element_blank(), ## element_line(colour = "white")
          panel.grid.minor=element_blank(), ## element_line(colour = "white", size = 0.25)
          axis.ticks=element_blank(),       ## element_line(colour = "grey20")
          axis.text=element_blank(),        ## element_text(size = rel(0.8), colour = "grey30")
          axis.title=element_blank(),      ## axis.title.x = element_text(
          plot.margin= grid::unit(c(0, 0, 0, 0), "in"),
          legend.position = 'None') -> prcp4

counties %>%
    left_join(weather %>%
              rename(county_fips = key) %>%
              filter(DATE == as.Date('2020-5-25'))) %>%
    ggplot(aes(fill = PRCP)) +
    geom_sf(size = 0) +
    scale_fill_material('blue') +  
    xlab('') + 
    ylab('') +
    ggtitle('May 25, 2020') + 
    labs(color = "Outcome") +
    theme_light() +
    theme(text = element_text(size=15),
          plot.title = element_text(vjust = -5, hjust = .75),
          strip.text.y = element_blank(),
          panel.border=element_blank(),
          panel.grid.major=element_blank(), ## element_line(colour = "white")
          panel.grid.minor=element_blank(), ## element_line(colour = "white", size = 0.25)
          axis.ticks=element_blank(),       ## element_line(colour = "grey20")
          axis.text=element_blank(),        ## element_text(size = rel(0.8), colour = "grey30")
          axis.title=element_blank(),      ## axis.title.x = element_text(
          plot.margin= grid::unit(c(0, 0, 0, 0), "in"),
          legend.position = 'None') -> prcp5

counties %>%
    left_join(weather %>%
              rename(county_fips = key) %>%
              filter(DATE == as.Date('2020-5-26'))) %>%
    ggplot(aes(fill = PRCP)) +
    geom_sf(size = 0) +
    scale_fill_material('blue') +  
    xlab('') + 
    ylab('') +
    ggtitle('May 26, 2020') + 
    labs(color = "Outcome") +
    theme_light() +
    theme(text = element_text(size=15),
          plot.title = element_text(vjust = -5, hjust = .75),
          strip.text.y = element_blank(),
          panel.border=element_blank(),
          panel.grid.major=element_blank(), ## element_line(colour = "white")
          panel.grid.minor=element_blank(), ## element_line(colour = "white", size = 0.25)
          axis.ticks=element_blank(),       ## element_line(colour = "grey20")
          axis.text=element_blank(),        ## element_text(size = rel(0.8), colour = "grey30")
          axis.title=element_blank(),      ## axis.title.x = element_text(
          plot.margin= grid::unit(c(0, 0, 0, 0), "in"),
          legend.position = 'None') -> prcp6
```
